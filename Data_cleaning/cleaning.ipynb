```python
def check(v):
    mois=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
    date=""
    ville=""
    for m in mois:
        if m in v:
            date=v[v.find(m):v.find(".")]
            ville=v.split(m)[0]
    return date,ville
```

def date_split(date):
    mois=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

        if m in date:
            temp=date.split("_")
            year=temp[1]
            month=m
            day=temp[0].split(m)[1]
    return year,month,day
