```python
import pandas as pd
import os
import numpy as np
```

```python
files_name= os.listdir("./Speeches")
```

```python
df=pd.DataFrame()
df["titre"]=files_name
```

```python
mois=["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
def take_ville(v):
    temp=""
    for m in mois:
        if m in v:
            temp=v.split(m)[0]
    return temp,m
```

```python
df["ville"]=None
df["date"]=None
```

```python

for i,v in enumerate(df["titre"]):
    temp=take_ville(v)
    df["ville"][i]=temp[0]

df
```

```python
def date_split(v):
    v=v.split(".")[0]
    for i in df["ville"]:
        if i in v:
            v=v.split(i)[1]
    return v
```

```python
for i,date in enumerate(df["titre"]):
    df.at[i,"date"]=date_split(date)
    
df
```

```python
df["content"]=None
for i,name in enumerate(df["titre"]):
    f=open(f"./Speeches/{name}","r",encoding="UTF-8")
    content=f.readlines()[0]
    f.close()
    df.at[i,"content"]=content

df["date"]=pd.to_datetime(df["date"],format="%b%d_%Y")
#df["date"]=pd.to_datetime(df["date"],format="%b%d_%Y")
    
df
    
```

```python
list=os.listdir("./")

try:
    df.to_csv("./clean_data/data.csv",sep=";")
except:
    os.mkdir("./clean_data/")
    df.to_csv("./clean_data/data.csv",sep=";")

```
